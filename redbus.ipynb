```python
import mysql.connector

# Establishing a connection to the MySQL database
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='7418736408',
    database='redbus',
    allow_local_infile=True
)

cursor = connection.cursor()

# Adjust the path to your CSV file
csv_file_path = "C:\\Users\\ANNIE CHRISTINA G\\Desktop\\REDBUS\\bus_data.csv"


# Creating the table if it does not exist
create_table_query = """
CREATE TABLE IF NOT EXISTS bus_routes (
    id INT AUTO_INCREMENT PRIMARY KEY,
    route_name VARCHAR(255),
    route_link VARCHAR(255),
    busname VARCHAR(255),
    bustype VARCHAR(255),
    departing_time TIME,
    duration TIME,
    reaching_time TIME,
    star_rating FLOAT,
    price DECIMAL(10, 2)
);
"""

cursor.execute(create_table_query)

# Constructing the LOAD DATA LOCAL INFILE query
load_data_query = f"""
LOAD DATA LOCAL INFILE '{csv_file_path}'
INTO TABLE bus_routes
FIELDS TERMINATED BY ','
ENCLOSED BY '"'
LINES TERMINATED BY '\n'
IGNORE 1 ROWS
(route_name, route_link, busname, bustype, departing_time, duration, reaching_time, star_rating, price);
"""

# Executing the LOAD DATA LOCAL INFILE query
try:
    cursor.execute(load_data_query)
    connection.commit()
    print("Data loaded successfully.")
except mysql.connector.Error as err:
    print(f"Error: {err}")

# Closing the cursor and connection
cursor.close()
connection.close()
```
